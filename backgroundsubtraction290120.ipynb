```python
from PIL import Image # (pip install Pillow)

import cv2
import os
from imutils import paths

path1='./smear2005D/'

directory= ['carcinoma_in_situ','light_dysplastic','moderate_dysplastic','severe_dysplastic','normal_columnar','normal_intermediate', 'normal_superficiel']

path2='./smear2005E/'

# path3='./smear2005D/'

basepath1=path1+directory[6]+'/'
basepath2=path2+directory[6]+'/'
# basepath3=path3+directory[6]+'/'



# PATH="/content/drive/My Drive/dataset_binary/dataset_binary/"
# basepath1= './dataset_binary_noback/smear2005D/normal/'
# folderpath= basepath+'abnormal/'
# basepath2 = './dataset_binary_noback/smear2005D/abnormal/'

# imagePaths = list(paths.list_images(folderpath))
# basepath3 = './dataset_binary_noback/smear2005D/normal/'

for filename in os.listdir(basepath1):
            
            path=basepath1+filename
            img=cv2.imread(path)  
            w,h=img.shape[:2]

            imag=img
            for i in range(w):
                for j in range(h):
                    if img[i,j,2]==255 and imag[i,j,0]==255 and imag[i,j,1]==255:
                    
                        imag[i,j,0]=0
                        imag[i,j,1]=0
                        imag[i,j,2]=0
            imag1=cv2.resize(imag,(256,256),interpolation = cv2.INTER_AREA)
            path2=basepath2+filename
            cv2.imwrite(path2,imag1)
            print(path2 +' is written')
```
